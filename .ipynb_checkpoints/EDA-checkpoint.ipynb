```python
import pandas as pd
import re

#Define a function to parse survey data 
def readDATFile(layout_file, dat_file):
    columns = []
    
    with open(layout_file, 'r') as file:
        for line in file:
           
            match = re.match(r'(\S+)\s+\S+\s+\S+\s+(\d+)\s+(\d+)', line)
            if match:
                col_name = match.group(1) 
                start_pos = int(match.group(2)) - 1  
                end_pos = int(match.group(3))  
                columns.append((col_name, start_pos, end_pos))
    colspecs = [(start, end) for _, start, end in columns]
    column_names = [name for name, _, _ in columns] 
    df = pd.read_fwf(dat_file, colspecs=colspecs, header=None, names=column_names) 
    return df 


survey2021 = readDATFile('pcg21Public/LAYOUTPCG21.TXT', 'pcg21Public/EPCG21.DAT')
```

```python
#Filter data to include only full-time, U.S.-based workers with a bachelor’s degree within certain age and work-hour limits, excluding specific employment types
survey2021 = survey2021[survey2021['BAIND'] == 'Y']
survey2021_filtered = survey2021[(survey2021['AGE'] >= 22) & (survey2021['AGE'] <= 65)] 
survey2021_filtered = survey2021_filtered[(survey2021_filtered['HRSWK'] >= 35) & (survey2021_filtered['HRSWK'] <= 85)]
survey2021_filtered  = survey2021_filtered[(survey2021_filtered ['ACFPT'] != 'Y')] 
survey2021_filtered = survey2021_filtered[(survey2021_filtered['EMTP'] != 16) & 
                          (survey2021_filtered['EMTP'] != 12) & 
                          (survey2021_filtered['EMUS'] == 'Y')]  
```

```python
survey2021_filtered
```

```python
# Convert BACARN to numeric, add "Degree_Location" based on degree origin
survey2021_filtered['BACARN'] = pd.to_numeric(survey2021_filtered['BACARN'])
survey2021_filtered['Degree_Location'] = survey2021_filtered['BADGRUS'].map({
    'Y': 'U.S.',
    'N': 'Non-U.S.'  
})
# Categorize institutions based on their academic focus
carnegie_mapping = {
    11: 'Research', 12: 'Research',
    13: 'Doctorate Granting', 14: 'Doctorate Granting',
    21: 'Comprehensive', 22: 'Comprehensive',
    31: 'Liberal Arts', 32: 'Liberal Arts',
    40: 'Specialized', 51: 'Specialized', 52: 'Specialized',
    53: 'Specialized', 54: 'Specialized', 56: 'Specialized',
    59: 'Specialized', 60: 'Specialized'
}

# Apply the mapping to create a grouped Carnegie classification column
survey2021_filtered['Carnegie_Group'] = survey2021_filtered['BACARN'].map(carnegie_mapping)

# Continue with the summary and pivot calculations
# Define a function to calculate demographic and field of study summaries by match type
def calculate_summary_by_match_type(df):
    # Calculate demographic percentages, field of study, and Carnegie group distribution by match type
    summary = df.groupby('OCEDRLP').agg(
        Men=('GENDER', lambda x: (x == 'M').mean() * 100),
        Black=('RACEM', lambda x: (x == 3).mean() * 100),
        Hispanic=('RACETHM', lambda x: (x == 4).mean() * 100),
        Married=('MARIND', lambda x: (x == 'Y').mean() * 100),
        # Degree level control variables
        Bachelor=('DGRDG', lambda x: (x == 1).mean() * 100),
        Master=('DGRDG', lambda x: (x == 2).mean() * 100),
        Doctorate=('DGRDG', lambda x: (x == 3).mean() * 100),
        Professional=('DGRDG', lambda x: (x == 4).mean() * 100),
        # Field of study based on NBAMEMG variable
        Comp_Math_Sci=('NBAMEMG', lambda x: (x == 1).mean() * 100),
        Bio_Agri_Life_Sci=('NBAMEMG', lambda x: (x == 2).mean() * 100),
        Phys_Sci=('NBAMEMG', lambda x: (x == 3).mean() * 100),
        Social_Sci=('NBAMEMG', lambda x: (x == 4).mean() * 100),
        Engineering=('NBAMEMG', lambda x: (x == 5).mean() * 100),
        S_E_Related=('NBAMEMG', lambda x: (x == 6).mean() * 100),
        Non_S_E=('NBAMEMG', lambda x: (x == 7).mean() * 100),
        # Carnegie Group classifications
        Research=('Carnegie_Group', lambda x: (x == 'Research').mean() * 100),
        Doctorate_Granting=('Carnegie_Group', lambda x: (x == 'Doctorate Granting').mean() * 100),
        Comprehensive=('Carnegie_Group', lambda x: (x == 'Comprehensive').mean() * 100),
        Liberal_Arts=('Carnegie_Group', lambda x: (x = 'Liberal Arts').mean() * 100),
        Specialized=('Carnegie_Group', lambda x: (x == 'Specialized').mean() * 100),
        # Degree Location
        US_Degree=('Degree_Location', lambda x: (x == 'U.S.').mean() * 100),
        Non_US_Degree=('Degree_Location', lambda x: (x == 'Non-U.S.').mean() * 100)
    ).reset_index()

    # Label match types for readability
    summary['Match_Type'] = summary['OCEDRLP'].replace({
        1: 'Match (Closely related)',
        2: 'Somewhat matched',
        3: 'Not related'
    })

    # Drop OCEDRLP column for clarity
    summary = summary.drop(columns=['OCEDRLP'])

    return summary

# Generate the summary table
table_df = calculate_summary_by_match_type(survey2021_filtered)

# Pivot the table so that Match_Type becomes columns instead of rows
table_pivoted = table_df.set_index('Match_Type').transpose()

# Rename columns for readability
table_pivoted.columns = ['Match (Closely related)', 'Somewhat Matched', 'Not Related']

# Display the final table formatted for a research paper
print("\nTable 4: Summary Statistics by Job Match Status (Match, Somewhat Matched, Not Related)")
print(table_pivoted)
```

```python
table_pivoted.to_csv("summary_table.csv")
```

```python
filtered_data = survey2021_filtered[(survey2021_filtered['SALARY'] >= 5000) & (survey2021_filtered['SALARY'] <= 500000)]

# Earnings summary statistics by Job Match Status after removing outliers
earnings_summary = filtered_data.groupby('OCEDRLP').agg(
    Earnings_Mean=('SALARY', 'mean'),
    Earnings_Median=('SALARY', 'median'),
    Earnings_STD=('SALARY', 'std'),
    Earnings_Min=('SALARY', 'min'),
    Earnings_25th_Percentile=('SALARY', lambda x: x.quantile(0.25)),
    Earnings_75th_Percentile=('SALARY', lambda x: x.quantile(0.75)),
    Earnings_Max=('SALARY', 'max')
)

# Transpose the table to switch rows and columns
earnings_summary_transposed = earnings_summary.transpose()

# Rename columns for readability based on job match types
earnings_summary_transposed.columns = ['Match (Closely related)', 'Somewhat Matched', 'Not Related']

# Display the final transposed table
print("Earnings Summary Statistics by Job Match Status (Outliers Removed)")
print(earnings_summary_transposed)
```

```python
earnings_summary_transposed.to_csv('salary_summary.csv')
```
